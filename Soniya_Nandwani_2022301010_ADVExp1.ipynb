```python
import pandas as pd

# Load the dataset
file_path = '/content/Sample - Superstore.csv'
df = pd.read_csv(file_path,encoding='ISO-8859-1')

# Display the first few rows of the dataframe
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart of Sales by Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title('Total Sales by Category')
plt.ylabel('Total Sales')
plt.xlabel('Category')
plt.show()
```

**Observation Of Bar Chart:** the chart shows the total sale by category. Among 3 categories:
Technology has highest total sales,surpassing the other categories. Office Supplies and Furniture have similar sales figures, with office supply slightly lagging behind Technology
Furniture ranks third in terms of total sales but still quite close to other categories.

```python
# Pie chart of Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 8))
region_sales.plot.pie(autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.ylabel('')
plt.show()
```

**Observation of Pie Chart:** Chart describes sales distribution across different regions: West holds largest share of total sales at 31.6%.
East holds closely with 29.5% of total sales.
Central accounts for 21.8% making it third largest.
South has smallest share,contributing 17.1% to total sales.

```python
# Histogram of Sales
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

**Observation of histogram:** The histogram shows that the majority of sales are concentrated in the lowest sales bracket, with a high frequency of low-value sales. Higher sales brackets have significantly lower frequencies, indicating that high-value sales are much less common. This suggests a skewed distribution where most sales transactions are of lower value.

```python
# Timeline chart of Sales over time
df['Order Date'] = pd.to_datetime(df['Order Date'])
sales_over_time = df.groupby('Order Date')['Sales'].sum()

plt.figure(figsize=(12, 6))
sales_over_time.plot()
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.show()
```

**Observation of TimeLine chart:** The chart shows sales over time from 2014 to 2018. Sales exhibit significant variability,with frequent spikes and drops. There are some peaks around mid-2014 and early 2016, where total sales reach above 20,000 units. Overall,the data suggests fluctuating sales without clear long term upward or downward pattern.

```python
# Scatter plot of Sales vs. Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

**Observation of Scatter plot:** the plot shows general positive correlation between sales and profit, with most data points clustered in the lower sales range(0-5000). However,there are outliers,including some instances of high sales with negative profits, few cases of very high profit at moderate to high sales level.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Alternative Bubble plot of Sales vs. Profit with bubble size by Discount and colored by Category
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Sales', y='Profit', size='Discount', hue='Category', data=df, sizes=(20, 200), alpha=0.6, palette='viridis')

plt.title('Sales vs. Profit (Bubble size by Discount, Colored by Category)')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Observation of buuble plot:** The plot shows the relationship between sales and profit across different product categories: Furniture (blue), Office Supplies (green), and Technology (purple). The size of the bubbles indicates the level of discount given, with larger bubbles representing higher discounts.From the chart, it is evident that as sales increase, profit generally tends to increase as well. However, this trend is not uniform across all categories. There are instances where higher sales do not necessarily lead to higher profits, especially when larger discounts are involved. Notably, some data points show negative profits despite having significant sales, suggesting that high discounts might be impacting profitability.
