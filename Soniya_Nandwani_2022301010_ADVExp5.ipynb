```r
# Install necessary packages (do this only once in the beginning)
install.packages("ggplot2")
install.packages("wordcloud")
install.packages("scatterplot3d")
install.packages("RColorBrewer")

# Load libraries (do this once in the beginning)
library(ggplot2)
library(wordcloud)
library(scatterplot3d)
library(RColorBrewer)

# Load the dataset
housing_data <- read.csv("/content/Housing.csv")
```

```r
# Word Cloud (Using text data in 'furnishingstatus')
text_data <- table(housing_data$furnishingstatus)
wordcloud(names(text_data), freq = text_data, scale = c(4, 0.5), colors = brewer.pal(8, "Dark2"))
```

**Observation of WordChart:** This colorful word cloud will show the most frequent categories of furnishing status, making the most common statuses stand out.

```r
# Box and Whisker Plot (Price vs Bedrooms)
ggplot(housing_data, aes(x = factor(bedrooms), y = price, fill = factor(bedrooms))) +
  geom_boxplot() +
  scale_fill_brewer(palette = "Set3") +  # Apply Set3 palette for color
  labs(title = "Boxplot of Price by Number of Bedrooms", x = "Number of Bedrooms", y = "Price") +
  theme_minimal()
```

**Observation of Box&Whisker plot:** As the number of bedrooms increases from 1 to 6, both the median price and the price range tend to rise. The boxplots show that properties with more bedrooms generally have higher prices, with significant variability and some outliers, especially in categories with 4 to 6 bedrooms. This suggests that larger properties are more expensive and have a wider range of prices.

```r
# Violin Plot (Area vs Stories)
ggplot(housing_data, aes(x = factor(stories), y = area, fill = factor(stories))) +
  geom_violin() +
  scale_fill_brewer(palette = "Pastel1") +  # Apply Pastel1 palette
  labs(title = "Violin Plot of Area by Number of Stories", x = "Number of Stories", y = "Area") +
  theme_minimal()
```

**Observation of Violin Plot:** shows the distribution of building areas for structures with 1 to 4 stories. Each violin shape represents the density of data points at different area values, with wider sections indicating higher frequencies. The plot reveals that buildings with more stories generally have larger areas, and the distribution of areas becomes more varied as the number of stories increases. This visualization helps in understanding how building area scales with the number of stories.

```r
# Linear Regression Plot (Price vs Area)
ggplot(housing_data, aes(x = area, y = price)) +
  geom_point(color = brewer.pal(8, "Blues")[6]) +  # Apply Blues palette for point color
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  labs(title = "Linear Regression: Price vs Area", x = "Area", y = "Price") +
  theme_minimal()
```

**Obervation of Linear Reg:** It shows a positive linear relationship between the area and price of properties. Each blue dot represents a data point, indicating individual properties, while the purple line represents the linear regression model. As the area increases, the price also tends to increase, suggesting that larger properties generally have higher prices. This plot is useful for predicting property prices based on their size and understanding market trends.

```r
# Nonlinear Regression Plot (Price vs Area)
ggplot(housing_data, aes(x = area, y = price)) +
  geom_point(color = brewer.pal(8, "Reds")[6]) +  # Apply Reds palette for point color
  geom_smooth(method = "loess", se = FALSE, color = "red") +
  labs(title = "Nonlinear Regression: Price vs Area", x = "Area", y = "Price") +
  theme_minimal()
```

**Observation of NonLinear Reg:** shows a nonlinear relationship between property area and price. The red dots represent individual data points, and the curve indicates that as the area increases, the price also increases, but not at a constant rate. This suggests that other factors might influence the price, or that the rate of price increase varies with different property sizes. This type of analysis is useful for understanding complex market trends and predicting property prices based on area.

```r
# 3D Scatter Plot
scatterplot3d(housing_data$area, housing_data$bedrooms, housing_data$price,
              color = "blue",  # Set color for points
              xlab = "Area", ylab = "Bedrooms", zlab = "Price",
              main = "3D Scatter Plot of Price, Area, and Bedrooms")
```

**Observation of Scatter plot:** It visualizes the relationship between property price, area, and the number of bedrooms. Each blue dot represents a property, showing how these three variables interact. The plot suggests that properties with larger areas and more bedrooms generally have higher prices. This type of visualization is useful for understanding complex relationships in real estate data and can help in making informed decisions based on multiple factors.

```r
# Jitter Plot (Price vs Parking)
ggplot(housing_data, aes(x = factor(parking), y = price)) +
  geom_jitter(width = 0.2, height = 0, color = brewer.pal(8, "Greens")[5]) +  # Apply Greens palette
  labs(title = "Jitter Plot of Price by Number of Parking Spaces", x = "Number of Parking Spaces", y = "Price") +
  theme_minimal()
```

**Observation of Jitter plot:**shows the distribution of property prices based on the number of parking spaces available. Each green dot represents a property, and the plot reveals three distinct clusters for 0, 1, and 2 parking spaces. Properties with more parking spaces generally have higher prices, but there is significant variability within each category. This visualization helps in understanding how parking availability might influence property prices, which can be useful for real estate market analysis.
